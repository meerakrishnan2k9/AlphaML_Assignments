```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.datasets import  fetch_california_housing
from sklearn.metrics import r2_score, mean_squared_error
```

```python
df=df.select_dtypes(include=np.number)
```

```python
print(df.shape)
```

```python
x = df[df.columns[0:17]]
print(x.shape)
```

```python
y = df[df.columns[-1]]
print(y.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20)
```

```python
print(x_train.shape, x_test.shape)
```

```python
w = []
r2 = []
mse=[]
alpha_list = [2]

for alpha in alpha_list:
    # fit model
    model = Lasso(alpha=alpha)
    model.fit(x_train, y_train)

    # prediction
    y_pred = model.predict(x_test)

    # store metrics for prediction, for each fit
    r2.append(r2_score(y_test, y_pred))
    mse.append(mean_squared_error(y_test, y_pred))
    w.append(model.coef_.tolist())
```

```python
# Visualise Feature Selection

plt.figure(figsize=(16, 10))
feature_names = list(x_train.columns)

for i in range(len(alpha_list)):
  plt.subplot(2, 2, i+1)
  plt.bar(feature_names, w[i])
  plt.title('Alpha = {}, r2 = {}'.format(alpha_list[i], r2[i]))

plt.show()
```

```python
# Visualise Feature Selection

plt.figure(figsize=(16, 10))
feature_names = list(x_train.columns)

for i in range(len(alpha_list)):
  plt.subplot(1, 1, i+1)
  plt.bar(feature_names, w[i])
  plt.title('Alpha = {}, mse = {}'.format(alpha_list[i], r2[i]))

plt.show()
```
